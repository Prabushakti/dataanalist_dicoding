### Cleaning Data

Membersihkan Tabel day_df

Menghapus kolom yang tidak dibutuhkan

```python
day_col_drop = ['windspeed', 'weekday']
day_df.drop(day_col_drop, axis=1, inplace=True)

day_df.head()
```

Column yang dihapus karena tidak diperlukan dalam analisa pertanyaan bisnis

Merubah penamaan kolom

```python
day_df.rename(columns={
    'dteday' : 'date',
    'yr' : 'year',
    'mnth' : 'month',
    'weathersit' : 'weather',
    'cnt' : 'total'
}, inplace=True)

day_df.head()
```

Mapping nilai kolom berupa angka menjadi keterangan

```python
#season
mapping_season = {
    1: 'springer', 2: 'summer',
    3: 'fall', 4: 'winter'
}
day_df['season'] = day_df['season'].map(mapping_season)

#tahun
mapping_year = {0: 2011, 1: 2012}
day_df['year'] = day_df['year'].map(mapping_year)

#bulan
mapping_month = {
    1: 'Jan', 2: 'Feb', 3: 'Mar',
    4: 'Apr', 5: 'May', 6: 'Jun',
    7: 'Jul', 8: 'Aug', 9: 'Sep',
    10: 'Oct', 11: 'Nov', 12: 'Dec'   
}
day_df['month'] = day_df['month'].map(mapping_month)

#weather
mapping_weather = {
    1: 'Clear', 2: 'Mist',
    3: 'Light Rain/Snow', 4: 'Heavy Rain/Snow'
}
day_df['weather'] = day_df['weather'].map(mapping_weather)


day_df.head()
```

```python
day_df.info()
```

```python
#merubah kolom ke datetime
day_df["date"] = pd.to_datetime(day_df["date"])

#merubah kolom ke category
day_df['season'] = day_df['season'].astype('category')
day_df['year'] = day_df['year'].astype('category')
day_df['month'] = day_df['month'].astype('category')
day_df['holiday'] = day_df['holiday'].astype('category')
day_df['workingday'] = day_df['workingday'].astype('category')
day_df['weather'] = day_df['weather'].astype('category')

day_df.info()
```

## Visualization & Explanatory Analysis

### Pertanyaan 1: Apa hubungan antara kondisi kelembaban dengan jumlah sewa sepeda per hari?

```python
plt.figure(figsize=(10, 6))
sns.set(style='whitegrid')
sns.lineplot(

    data=day_df, 
    x='hum',  
    y='total',  
    marker='o'
)
plt.title("Hubungan Antara Kondisi Kelembaban dan Jumlah Sewa Sepeda Perhari")
plt.xlabel("Kelembaban Udara ")
plt.ylabel("Jumlah Sewa")
plt.show()
```

### Pertanyaan 2: Sepeda paling banyak dan paling sedikit direntalkan pada musim apa saja?

```python
plot_season = day_df.groupby('season', observed=False)[['registered', 'casual']].sum().reset_index()

plt.figure(figsize=(10, 6))
sns.barplot(

    data=plot_season, 
    x='season', 
    y='registered', 
    label='Registered', 
    color='tab:blue'
)

sns.barplot(
    
    data=plot_season, 
    x='season', 
    y='casual', 
    label='Casual', 
    color='tab:orange'
)

plt.title('Sepeda yang Paling Banyak dan Paling Sedikit Direntalkan Setiap Musim')
plt.xlabel(None)
plt.ylabel(None)
plt.legend()
plt.show()
```

total penyewaan sepeda paling banyak terjadi pada musim fall baik untuk pelanggan casual maupun pelanggan yang sudah registered. Sebaliknya total penyewaaan sepeda paling sedikit terjadi pada musing springer.

### Pertanyaan 3 : Bagaimana kinerja rental sepeda setiap bulan antara tahun 2011 hingga 2012?

```python
plot_month = day_df['month'].astype(str)
plot_year = day_df['year'].astype(str)

day_df['year_month'] = plot_month + ' ' + plot_year

day_df['total_sum'] = day_df.groupby('year_month')['total'].transform('sum')

plt.figure(figsize=(20, 6))
sns.set(style='whitegrid')
sns.lineplot(
    data=day_df, 
    x='year_month', 
    y='total_sum',  
    marker='o'
)
plt.title("Total Sepeda yang Direntalkan Setiap Bulan Antara Tahun 2011 Hingga 2012")
plt.xlabel(None)
plt.ylabel(None)
plt.xticks(rotation=45)
plt.show()
```

terjadi peningkatan minat penyewaan sepeda yang cukup tinggi di tahun 2012 apabila dibandingkan dengan tahun sebelumnya.

B. Conclution pertanyaan 2 :

Musim yang paling sedikit memiliki sewa sepeda adalah musim spring (musim semi), dengan bar berwarna biru yang paling rendah.

    Sementara itu, untuk sewa sepeda secara tidak terdaftar (casual), musim yang paling banyak adalah summer (musim panas), ditunjukkan oleh bar berwarna oranye yang tertinggi, sementara musim yang paling sedikit adalah winter (musim dingin), ditunjukkan oleh bar berwarna oranye yang paling rendah.

    Jadi, secara umum, musim panas (summer) merupakan musim di mana terjadi peningkatan signifikan dalam sewa sepeda, baik yang terdaftar maupun tidak terdaftar. Sedangkan musim semi (spring) adalah musim dengan jumlah sewa sepeda yang paling rendah.

C. Conclution pertanyaan 3 :

Pada awal tahun 2012, tren ini berlanjut dengan jumlah sewa sepeda yang terus meningkat secara konsisten hingga pertengahan tahun, di mana mencapai puncaknya. Setelah itu, terjadi sedikit penurunan hingga akhir tahun 2012, tetapi jumlahnya tetap relatif tinggi dibandingkan dengan awal tahun 2011.
